# Clearning

```python
backup = df.copy()
```

#### Year is not in INT and has non INT values.

```python
# Keep the INT Values and remove the NON INT values which are in String Format
df = df[df['year'].str.isnumeric()]
```

#### Year is not in INT Datatype.

```python
# changing the Datatype of Year from Object to int
df['year'] = df['year'].astype(int)
```

#### Price has non INT value :- Ask for Price

```python
# Removing "Ask for Price"
df = df[df.Price != "Ask For Price"]
```

#### Price has Commas

```python
# Replace the Commas with Empty String
df['Price'] = df['Price'].str.replace(',', '')
```

#### Price is in Object and not in INT

```python
df['Price'] = df['Price'].astype(int)
```

#### kms_driven contains :- Petrol, kms, Commas so remove kms and Commas then remove Petrol

```python
# Remove kms and Commas
df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
```

```python
# Remove Petrol
df = df[df['kms_driven'].str.isnumeric()]
```

```python
# Conver to INT
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
# Replace NULL Values with the mean of that Column
df['kms_driven'] = df['kms_driven'].fillna(df['kms_driven'].mean())
```

#### fule_type has a NAN/NULL Value

```python
df = df[~df['fuel_type'].isna()]
```

```python
df['fuel_type'].isna().sum()
```

#### Take upto first 3 or if not available then >3 Words in NAME

```python
df['name'] = df['name'].str.split(' ').str.slice(0, 3).str.join(' ')
```

```python
# It will bring a new Index Column and let the previous Index Column as it is with the name as "index" but :- drop = True will remove the previous index column
df = df.reset_index(drop = True)
```

# Model

```python
x = df.drop(columns = 'Price', axis = 1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

### Used for Label Encoding of Columns

We passed all the categories or data to OHE and not just x_train rather the whole dataset

```python
ohe = OneHotEncoder()
ohe.fit(x[['name', 'company', 'fuel_type']])
# FIT is done to make available these columns as categories of 'OHE'
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_), ['name', 'company', 'fuel_type']), remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

PIPE is used because when we input the data from website so we have to write a diff function to LABEL ENCODE and feed it to the LinearRegression :- So we created a PIPE in which we have stored the ENCODER and LR on one side and from the other side the Predictions are going to be out.

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_predict = pipe.predict(x_test)
```

```python
r2_score(y_test, y_predict)
```

Now we want to get the MAX R2 score.

```python
scores = []
for i in range(1000):
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = i)
  pipe = make_pipeline(column_trans, lr)
  pipe.fit(x_train, y_train)
  y_predict = pipe.predict(x_test)
  scores.append(r2_score(y_test, y_predict))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = np.argmax(scores))
pipe = make_pipeline(column_trans, lr)
pipe.fit(x_train, y_train)
y_predict = pipe.predict(x_test)
r2_score(y_test, y_predict)
```
